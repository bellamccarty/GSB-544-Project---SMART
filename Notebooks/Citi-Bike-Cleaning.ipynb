#### Below are extracted citi-bike trip data from 2023

```python
import pandas as pd
import sns
from matplotlib import pyplot as plt
```

```python
import pandas as pd
import os

def read_first_citibike_data(base_folder_path):
    """
    Reads the first CSV file from each monthly CitiBike trip data subfolder (Jan, Feb, ..., Dec)
    under the specified base folder path and combines them into a single DataFrame.

    Parameters:
    base_folder_path (str): The base path containing subfolders for each month (e.g., Jan, Feb).

    Returns:
    pd.DataFrame: A DataFrame containing data from the first CSV file in each monthly folder.
    """
    combined_df = pd.DataFrame()  # Initialize an empty DataFrame
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']  # List of month abbreviations

    # Loop through each month subfolder
    for month in months:
        month_folder_path = os.path.join(base_folder_path, f"{month}-citibike-tripdata")
        if os.path.exists(month_folder_path):  # Check if the month folder exists
            # List all CSV files in the month folder
            csv_files = [file for file in os.listdir(month_folder_path) if file.endswith('.csv')]
            if csv_files:  # Check if there are any CSV files
                # Sort to ensure consistent order if needed (optional, based on your requirements)
                csv_files.sort()  # This can be omitted or customized based on file naming patterns
                first_file_path = os.path.join(month_folder_path, csv_files[0])  # Get the first file
                # Read the first CSV into a DataFrame
                temp_df = pd.read_csv(first_file_path)
                # Optionally, add a column with the month for reference
                temp_df['month'] = month
                # Append the DataFrame
                combined_df = pd.concat([combined_df, temp_df], ignore_index=True)

    return combined_df

# Example usage:
base_folder_path = r"C:\Users\isabe\PycharmProjects\GSB-544-Project---SMART\Data\CitiBike_TripData"
combined_data = read_first_citibike_data(base_folder_path)
print(combined_data.head())  # Display the first few rows of the combined DataFrame
```

```python
import pandas as pd

# Step 1: Load CitiBike data (assuming combined_data is your DataFrame)
combined_data['started_at'] = pd.to_datetime(combined_data['started_at'])
combined_data['date'] = combined_data['started_at'].dt.date  # Extract date part

# Convert columns to appropriate data types
combined_data['ride_id'] = combined_data['ride_id'].astype(str)
combined_data['rideable_type'] = combined_data['rideable_type'].astype('category')
combined_data['start_station_name'] = combined_data['start_station_name'].astype('category')
combined_data['start_station_id'] = pd.to_numeric(combined_data['start_station_id'], errors='coerce')
combined_data['end_station_name'] = combined_data['end_station_name'].astype('category')
combined_data['end_station_id'] = pd.to_numeric(combined_data['end_station_id'], errors='coerce')
combined_data['start_lat'] = pd.to_numeric(combined_data['start_lat'], errors='coerce')
combined_data['start_lng'] = pd.to_numeric(combined_data['start_lng'], errors='coerce')
combined_data['end_lat'] = pd.to_numeric(combined_data['end_lat'], errors='coerce')
combined_data['end_lng'] = pd.to_numeric(combined_data['end_lng'], errors='coerce')
combined_data['member_casual'] = combined_data['member_casual'].astype('category')
combined_data['month'] = combined_data['month'].astype('category')

# Step 2: Load weather data
weather_data = pd.read_csv(r'C:\Users\isabe\PycharmProjects\GSB-544-Project---SMART\Data\central_park_weather.csv')
weather_data['date'] = pd.to_datetime(weather_data['DATE']).dt.date  # Convert date column to date

# Step 3: Merge the two datasets on the date column
merged_data = pd.merge(combined_data, weather_data, on='date', how='left')

# Display the merged data
merged_data.head()
```

```python
merged_data['rideable_type'].value_counts().plot(kind='bar', figsize=(8, 5))
plt.title('Distribution of Rides by Rideable Type')
plt.xlabel('Rideable Type')
plt.ylabel('Ride Count')
plt.show()
```

```python
import numpy as np

heatmap_data = merged_data.groupby([pd.cut(merged_data['TMAX'], bins=np.arange(30, 101, 10)), 
                                    pd.cut(merged_data['PRCP'], bins=np.linspace(0, 1, 6))]).size().unstack()
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt="d")
plt.title('Heatmap of Ride Counts by Temperature and Precipitation')
plt.xlabel('Precipitation (inches)')
plt.ylabel('Max Temperature (Â°F)')
plt.show()
```

```python
merged_data['trip_duration'] = (pd.to_datetime(merged_data['ended_at']) - pd.to_datetime(merged_data['started_at'])).dt.total_seconds() / 60  # Convert to minutes
merged_data['temp_bins'] = pd.cut(merged_data['TMAX'], bins=np.arange(30, 101, 10))

plt.figure(figsize=(12, 6))
sns.boxplot(data=merged_data, x='temp_bins', y='trip_duration')
plt.title('Trip Duration Distribution vs. Max Temperature')
plt.xlabel('Max Temperature (Â°F)')
plt.ylabel('Trip Duration (minutes)')
plt.xticks(rotation=45)
plt.show()
```
